Threshold

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('img/MicrosoftTeams-image3.png')
gRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(gRGB, cv.COLOR_RGB2GRAY)
plt.hist(gray.ravel(), 256, [0, 256]); plt.show()
ret, thresh1 = cv.threshold(gray, 100, 255, cv.THRESH_BINARY)

# print
plt.imshow(gRGB)
plt.figure()
plt.imshow(gray, cmap='gray')
plt.figure()
plt.imshow(thresh1, cmap='gray')
```

Histogram Equalization

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('img/MicrosoftTeams-image1.png')
gRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(gRGB, cv.COLOR_RGB2GRAY)
plt.hist(gray.ravel(), 256, [0, 256]); plt.show()
hieq = cv.equalizeHist(gray)

# print
plt.imshow(gRGB)
plt.figure()
plt.imshow(gray, cmap='gray')
plt.figure()
plt.imshow(hieq, cmap='gray')
plt.figure()

plt.hist(hieq.ravel(), 256, [0, 256]); plt.show()
```
