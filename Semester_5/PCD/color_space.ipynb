```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('img/MicrosoftTeams-image3.png')
plt.figure()
plt.imshow(img)
plt.figure()

for (name, chan) in zip(("B", "G", "R"), cv.split(img)):
    plt.imshow(chan, cmap='gray')
```
