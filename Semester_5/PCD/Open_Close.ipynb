```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img= cv.imread('img/bersihkan noise.png')
img= cv.imread('img/hitung_koin.png')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)

kernel = np.ones((5,5), np.uint8)
ero = cv.erode(img, kernel, iterations=1)
dil = cv.dilate(img, kernel, iterations=1)
open = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
close = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.title("Source"); plt.imshow(img_rgb); plt.figure()
plt.title("Erosion"); plt.imshow(ero); plt.figure()
plt.title("Dilation"); plt.imshow(dil); plt.figure()
plt.title("Opening"); plt.imshow(open); plt.figure()
plt.title("Closing"); plt.imshow(close); plt.figure()
```
